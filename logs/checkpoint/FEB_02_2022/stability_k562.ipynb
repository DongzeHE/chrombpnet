```python
import h5py
import matplotlib.pyplot as plt
idx=500



```

```python
model1 = "/mnt/lab_data2/anusri/chrombpnet/results/chrombpnet/ATAC_PE/K562_stability/K562_02.08.2022_bias_128_4_1234/bias_model/bias_predictions.h5"
model2 = "/mnt/lab_data2/anusri/chrombpnet/results/chrombpnet/ATAC_PE/K562_stability/K562_02.08.2022_bias_128_4_2356/bias_model/bias_predictions.h5"
model3 = "/mnt/lab_data2/anusri/chrombpnet/results/chrombpnet/ATAC_PE/K562_stability/K562_02.08.2022_bias_128_4_5678/bias_model/bias_predictions.h5"
```
