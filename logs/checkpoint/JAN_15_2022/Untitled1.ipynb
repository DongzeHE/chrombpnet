```python
import deepdish
bias = deepdish.io.load(folder+"bias_footprints.h5")
corrected = deepdish.io.load(folder+"corrected_footprints.h5")
```

```python
bias.keys()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,10)
fig, axs = plt.subplots(2,2, sharex=True, sharey=False)
ylims=[0.01,0.01,0.006,0.002,0.002,0.004]

start=500-100+5
end=500+100+5
width = end - start


i=0
for key in bias:
    if "CTACF" in key:
        continue
    else:
        ylim=ylims[0]
        ax=axs[i,0]
        ax.plot(range(width), bias[key][0][start:end], 'k', linewidth=1, alpha=0.5)
        ax.set_ylim(0,ylim)
    
        
        ylim=ylims[0]
        ax=axs[i,1]
        ax.plot(range(width), corrected[key][0][start:end], 'k', linewidth=1, alpha=0.5)
        #ax.set_ylim(0,ylim)
        i+=1
        

fig.subplots_adjust(wspace=0.5, hspace=0.2)

plt.show()
```
