```r
library(ggseqlogo)
library(ggplot2)
library(patchwork)
```

```r
read_nptxt_matrix <- function(fname) {
    d=read.table(fname)
    colnames(d) = c("A", "C", "G", "T")
    t(d)
}
```

```r

subcluster_base_path="addons/counts_0_0.pfm"
submotif = read_nptxt_matrix(subcluster_base_path)
```

```r
m=submotif

revcomp <- function(x) {
    x = x[rev(seq(nrow(x))), rev(seq(ncol(x)))]
    rownames(x) = c("A", "C", "G", "T")
    x
}

m = revcomp(m)                

options(repr.plot.width = 8, repr.plot.height = 1)

rownames(m) = c("A","C","G","T")
    
f="helvetica_regular"
    
p <- ggseqlogo(m, method="custom", font=f, ncol=1, size=12) + 
                                    theme_void() 
    
p

ggsave("subfigs/gata_tal_motif.pdf", device = "pdf",  width = 8, height = 1, dpi = 300)
```
